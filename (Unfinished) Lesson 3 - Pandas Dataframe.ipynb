## Data Manipulations with Pandas

Sometimes we would like to manipulate datasets for further processing. For example, feature engineering and data cleaning, which clean data and add additional features for easier analysis. Pandas also provide some features that help us to do so, let's go through some of them.

From the dataset above, we might interest in the rank of the countries, but sometimes we interpret rank with different meaning. That can be decided by the number of gold medals won by the country or we just count their total number of medals. There's ambiguity for measurements. 

Here, we can do sorting with different columns to see how's the ranking varies.

In python, a self-defined functions can be written as follow:

> `def function_name(param1, param2,...): # and your code below`

In simple words, you can use function (or called definition) for some sequence of operations you are trying to perform multipled times. For example here, we might want to give the continent name to each record. 

```python
## Credits to Isra @ Stackoverflow 
## https://stackoverflow.com/questions/55910004/get-continent-name-from-country-using-pycountry
import pycountry_convert as pc

def country_to_continent(country_name):
    try:
      country_alpha2 = pc.country_name_to_country_alpha2(country_name)
      country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
      country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)  
    except:
      country_continent_name = '.'
    return country_continent_name

# Example
country_name = input("Enter a Country name: ")
print(country_to_continent(country_name))
```

We can use `dataframe.apply(function_name)` to give a new column like below, we can look into country that the function can't detect and give output.

```python
dataframe['Continent'] = dataframe['Country'].apply(country_to_continent)

dataframe[dataframe['Continent'] == '.']
```
