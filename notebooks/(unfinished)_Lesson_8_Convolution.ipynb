```python
import os
import zipfile

local_zip = '/content/cats_and_dogs_filtered.zip'

zip_ref = zipfile.ZipFile(local_zip, 'r')

zip_ref.extractall('/content')
zip_ref.close()
```

## Setting Directory

```python
base_dir = '/content/cats_and_dogs_filtered'

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')

train_cats_dir = os.path.join(train_dir, 'cats')
train_dogs_dir = os.path.join(train_dir, 'dogs')

validation_cats_dir = os.path.join(validation_dir, 'cats')
validation_dogs_dir = os.path.join(validation_dir, 'dogs')
```

```python
print(train_dogs_dir)
print(train_cats_dir)
print(validation_cats_dir)
print(validation_dogs_dir)
```

```python
print("Number of images in cat training dir: ", len(os.listdir(train_cats_dir)))
print("Number of images in dog training dir: ", len(os.listdir(train_dogs_dir)))
print("Number of images in cat validation dir: ", len(os.listdir(validation_cats_dir)))
print("Number of images in dog validation dir: ", len(os.listdir(validation_dogs_dir)))
```
